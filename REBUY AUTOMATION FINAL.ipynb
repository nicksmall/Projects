## Calculate BuyBox Prices

```python
def buy_box_price(row):
    if row['new_third_party_fba'] > row['new_low_price_fba'] and row['buybox_new_landed_price'] == 0 or pd.isnull(row['buybox_new_landed_price']) :
        return row['new_low_price_fba']
    elif row['new_third_party_fba'] < row['new_low_price_fba'] and row['buybox_new_landed_price'] == 0 or pd.isnull(row['buybox_new_landed_price']) :
        return row['new_third_party_fba']
    elif  row['buybox_new_landed_price'] == 0 and pd.notnull(row['new_third_party_fba']) and  pd.isnull(row['new_low_price_fba']):
        return row['new_third_party_fba']
    
    elif  row['buybox_new_landed_price'] == 0 and pd.isnull(row['new_third_party_fba']) and  pd.notnull(row['new_low_price_fba']):
        return row['new_low_price_fba']
    elif pd.notnull(row['new_third_party_fba'] ) and pd.isnull(row['new_low_price_fba']) and pd.isnull(row['buybox_new_landed_price']):
        return row['new_third_party_fba']
    

    
    elif pd.isnull(row['new_third_party_fba'] ) and pd.notnull(row['new_low_price_fba']) and pd.isnull(row['buybox_new_landed_price']):
        return row['new_low_price_fba']

    
    
    
    
    
    elif pd.notnull(row['buybox_new_landed_price']):
        #return row['buybox_new_landed_price']
        return row['buybox_new_landed_price']
   

    
df['BB_PRICE'] = df.apply(buy_box_price, axis=1)
```

## Flag Empty Buy Box Price

```python
def flag_price_empty(row):
    if pd.isnull(row['new_third_party_fba']) and pd.isnull(row['new_low_price_fba']) and pd.isnull(row['buybox_new_landed_price']):
        return 'no price values'
    elif pd.isnull(row['new_third_party_fba']) and pd.isnull(row['new_low_price_fba']) and row['buybox_new_landed_price'] == 0:
        return 'price is zero'
    elif row['new_third_party_fba'] == 0 or row['new_low_price_fba'] == 0 or row['buybox_new_landed_price'] == 0:
        return 'price is zero'
    
    
    
df['flag_BB_PRICE'] = df.apply(flag_price_empty, axis=1)
```

## Calculates List Pack from SKU

```python
# LP = list pack (if pack of 6, LP =6, if single, LP = 1)

# LP comes from the sku, it comes from the last 3 digits - 100
df['skulen'] = df['SKU'].astype(str).map(len)
print(df['skulen'].value_counts())
df['SKU'].loc[(df.skulen == 21)]

#ignore dash, just take sku 19 and 20 <-- this not true - ask michael 

#get last item in string of SKU, put it in a list, and then just get the unique values 
sku_endings = list(set(df['SKU'].str[-1:].tolist()))


# convert numbers to float while leaving strings in string format b/c want a list of just strings 

def maybe_float(s):
    try:
        return float(s)
    except (ValueError, TypeError):
        return s


sku_endings = [maybe_float(v) for v in sku_endings]


#get just the string endings 
sku_endings_str = [ x for x in sku_endings if type(x)==str ]


def f(row):
    if row['skulen'] == 19:
        return row['SKU'][16:]
    elif row['skulen'] == 20 and row['SKU'].endswith(tuple(sku_endings_str)):
        return row['SKU'][16:19]
    
    elif row['skulen'] == 20 :
        return row['SKU'][17:20]
    
df['LP'] = df.apply(f,axis=1)
```

## Converts LP to Float and then converts it to single digits by subtracting 100, 200, or 300.

```python
df['LP'] = df['LP'].astype(float)

df.loc[(df['LP'] >= 100) & (df['LP'] < 200), 'LP'] = df['LP'] - 100
df.loc[(df['LP'] >= 200) & (df['LP'] < 300), 'LP'] = df['LP'] - 200
df.loc[(df['LP'] >= 300) & (df['LP'] < 400), 'LP'] = df['LP'] - 300
df.loc[(df['LP'] >= 800) & (df['LP'] < 900), 'LP'] = df['LP'] - 800
```
